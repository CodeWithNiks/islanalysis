LEFT-FOOT VS RIGHT-FOOT PLAYERS

```python
n_left_footed_players = isl7["player_foot"].value_counts()["left"]
n_right_footed_players = isl7["player_foot"].value_counts()["right"]


plt.pie(
    [n_left_footed_players, n_right_footed_players],
    labels=["Left-footed", "Right-footed"],
    autopct="%1.1f%%",
)
plt.title("Pie chart of left-footed and right-footed players")
plt.show()
```

PLAYERS OF DIFFERENT COUNTRIES

```python


n_players_by_country = df["country_name"].value_counts().sort_values(ascending=False)

plt.pie(
    n_players_by_country.values,
    labels=n_players_by_country.index,
    autopct="%1.1f%%",
)
plt.title("Pie chart of number of players by country")
plt.show()
```

FINDING PLAYERS OF HEIGHT LEVEL

```python
column = "height"


isl7_filtered = isl7[isl7[column] >= 150]


isl7_filtered = isl7_filtered.dropna()

grouped_data = isl7_filtered.groupby(column)

n_player_per_height = grouped_data["name"].count()

plt.bar(n_player_per_height.index, n_player_per_height.values)
plt.xlabel("Height (cm)")
plt.ylabel("Number of players")
plt.title("Number of players of different heights ")
plt.show()
```

LIST OF TOP 10 GOAL SCORERS

```python

columns = ["name", "events.goals"]


df_sorted = df.sort_values(by="events.goals", ascending=False)

top_10_goalscorers = df_sorted[0:10]

plt.figure(figsize=(10, 6))
sns.barplot(
    x="name",
    y="events.goals",
    data=top_10_goalscorers,
    color="green",
    edgecolor="black",
    linewidth=1,
    width=0.2
)
plt.xlabel("Player name")
plt.ylabel("Number of goals scored")
plt.title("Top 10 goalscorers")
plt.show()
```

PREDICTING TOP 10 VALUABLE FORWARDS IN NEXT SEASON

```python
columns = ["events.goals", "events.assists", "events.key_passes", "events.chances_created", "events.shots_on_target", "touches.successful_tackles"]


isl7["shot_conversion_rate"] = isl7["events.goals"] / isl7["events.shots"]



model = LinearRegression()


model.fit(isl7[columns], isl7["events.goals"])


predictions = model.predict(isl7[columns])
print(mean_squared_error(isl7["events.goals"], predictions))


isl7 = isl7.assign(next_season_predictions=predictions)
next_season_predictions = model.predict(isl7[columns])

isl7_sorted = isl7.sort_values(by="next_season_predictions", ascending=False)
print(isl7_sorted[0:10])
```
